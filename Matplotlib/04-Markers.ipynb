```python
import matplotlib.pyplot as plt
import numpy as np
```

# Marker Reference

## You can choose any of these markers:
Marker 	Description
- 'o' 	Circle 	
- '*' 	Star 	
- '.' 	Point 	
- ',' 	Pixel 	
- 'x' 	X 	
- 'X' 	X (filled) 	
- '+' 	Plus 	
- 'P' 	Plus (filled) 	
- 's' 	Square 	
- 'D' 	Diamond 	
- 'd' 	Diamond (thin) 	
- 'p' 	Pentagon 	
- 'H' 	Hexagon 	
- 'h' 	Hexagon 	
- 'v' 	Triangle Down 	
- '^' 	Triangle Up 	
- '<' 	Triangle Left 	
- '>' 	Triangle Right 	
- '1' 	Tri Down 	
- '2' 	Tri Up 	
- '3' 	Tri Left 	
- '4' 	Tri Right 	
- '|' 	Vline 	
- '_' 	Hline

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker="o")
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker="_")
plt.show()
```

# Format Strings 'fmt'
## Syntax = marker|line|color

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, 'o:g')
plt.show()
```

# Line Reference

Line Syntax 	Description

- '-' 	Solid line 	
- ':' 	Dotted line 	
- '--' 	Dashed line 	
- '-.' 	Dashed/dotted line

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, "_-k")
plt.show()
```

# Marker Size

```python
ypoints = np.array([3, 8, 1, 10])

# ms = marker Size / mec = marker color /mfc =marker face(inside color)
plt.plot(ypoints, marker = 'o', ms = 20, mec='r', mfc = 'k')
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker="h", ms=20, mec="#4CAF50", mfc="#4CAF50")
plt.show()
```
