```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

# the figure has 1 row, 2 columns, and this plot is the first plot.
plt.subplot(1, 2, 1, title='main Plot')
plt.plot(x, y)

# plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

# the figure has 1 row, 2 columns, and this plot is the second plot.
plt.subplot(1, 2, 2, title='not main plot')
plt.plot(x, y)

plt.suptitle('my whole plot')
plt.show()
```
