```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
xpoints = np.array([1, 3])
ypoints = np.array([8, 10])

# takes parameters 'o' means rings
plt.plot(xpoints, ypoints, 'o')
plt.show()
```
