```python
import pandas as pd
```

```python
a = [1, 2 ,3]

s = pd.Series(a)

print(s[2])
```

```python
a = [1, 2 ,3]

s = pd.Series(a, index=['x', 'y', 'z'])

print(s)
print(s['x'])
```

```python
calories = {
    'day1' : 420,
    'day2' : 500,
    'day3' : 320
}

s = pd.Series(calories)

print(s)
```

```python
# you can take custom data

calories = {
    'day1' : 420,
    'day2' : 500,
    'day3' : 320
}

s = pd.Series(calories, index=['day1', 'day3'])

print(s)
```

```python
data = {
    'calories' : [300, 400, 500],
    "duration": [50, 40, 45]
}

myframe = pd.DataFrame(data)

print(myframe)
```
