```python
import numpy as np
```

```python
arr = np.array([3, 2, 0, 1])

#NOTE: Creates a Copy
arr_sorted = np.sort(arr)

print(arr_sorted)
```

```python
arr = np.array(['banana', 'cherry', 'apple'])
 
arr_sorted = np.sort(arr)

print(arr_sorted)
```

```python
arr = np.array([True, False, True])

print(np.sort(arr))
```

```python
arr = np.array([[3, 2, 4], [5, 0, 1]])

#NOTE: Sorts Both
print(np.sort(arr))
```
