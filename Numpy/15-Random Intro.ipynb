# Generating Random Numbers

```python
from numpy import random
```

```python

x = random.randint(100)

print(x)
```

# Random Float

```python
x = random.rand() 

print(x)
```

# Random Arrays

```python
x = random.randint(100, size=5)

print(x)
```

```python
x = random.randint(100, size=(5, 3)) # Returns a 5x3 matrix

print(x)
```

```python
x = random.rand(5)

print(x)
```

```python
x = random.rand(3, 5)

print(x)
```

```python
#NOTE: size is the number of choices
x = random.choice([1, 2, 3, 4], size=2)

print(x)
```

```python
x = random.choice([3, 5, 7, 9], size=(3, 5))
print(x)
```
