```python
import numpy as np
```

```python
nums = [2, 3, 4, 5]
```

```python
arr = np.array(nums)
```

## 0-D (Zero Dimensional)

```python
arr1 = np.array(42)

print(arr) 
```

## 1-D (One Dimensional)

```python
arr2 = np.array([1, 2, 3, 4])

print(arr2)
```

## 2-D (Two Dimensional)

```python
arr3 = np.array([[1, 2, 3, 4], [5, 3, 1, 5]])

print(arr3)
```

## 3-D (Three Dimensional)

```python

arr = np.array([[[1, 2, 3], [4, 5, 6]], [[1, 2, 3], [4, 5, 6]]])

print(arr) 
```

# Checking Number of Dimentions:

```python
print(arr.ndim) #? Number of dimentions
print(arr1.ndim)
print(arr2.ndim)
print(arr3.ndim)
```

```python
arr5 = np.array([1, 2, 3, 4], ndmin=5)

print(arr5)
print(arr5.ndim)
```
