```python
import numpy as np
```

```python
arr = np.array([1, 2, 3, 4])

arrCopied = arr.copy()

arr[0] = 100

print(arr ,'\n', arrCopied)
```

```python
arr = np.array([1, 2, 3, 4])

arrCopied = arr.view()

arr[0] = 100

print(arr, "\n", arrCopied)
```

```python
arr = np.array([1, 2, 3, 4])

arrViewed = arr.view()

arrViewed[0] = 100

print(arr, "\n", arrCopied)
```
