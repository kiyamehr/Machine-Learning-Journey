```python
import numpy as np
```

# Joining to Arrays Together using concatenate

```python
arr1 = np.array([1, 2, 3])

arr2 = np.array([4, 5, 6])

arrJoined = np.concatenate((arr1, arr2))

print(arrJoined)
```

```python
arr1 = np.array([[1, 2], [3, 4]])

arr2 = np.array([[5, 6], [7, 8]])

arrJoined = np.concatenate((arr1, arr2), axis=1)

print(arrJoined)
```

```python
arr1 = np.array([1, 2, 3])

arr2 = np.array([4, 5, 6])

arr_stacked = np.stack((arr1, arr2), axis=0)

print(arr_stacked)
```

```python
arr1 = np.array([1, 2, 3])

arr2 = np.array([4, 5, 6])

arr = np.hstack((arr1, arr2))

print(arr)
```

```python
arr1 = np.array([1, 2, 3])

arr2 = np.array([4, 5, 6])

arr = np.vstack((arr1, arr2))

print(arr)
```

```python
arr1 = np.array([1, 2, 3])

arr2 = np.array([4, 5, 6])

arr = np.dstack((arr1, arr2))

print(arr)
```
