```python
import numpy as np
import sklearn as skl
from sklearn.linear_model import LinearRegression
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def remove_nan(dataset):
    cleaned = dataset[~np.isnan(dataset)]
    return cleaned
```

```python
data = pd.read_csv('./Data/houses.csv')
# print(data.to_string())
```

```python
data_numpy = np.genfromtxt("./Data/houses.csv", delimiter=",")

print(data_numpy)
```

```python
x_train = np.take(data_numpy, indices=0,axis=1)
x_train = remove_nan(x_train).reshape(-1, 1)


print(x_train)
```

```python
y_train = np.take(data_numpy, indices=1, axis=1)
y_train = remove_nan(y_train).reshape(-1, 1)
print(y_train)
```

```python
plt.scatter(x_train, y_train)
plt.show()
```

```python
regr = LinearRegression()
regr.fit(x_train, y_train)

print(regr.score(x_train, y_train))
```

```python
y_predict = regr.predict(x_train)

plt.scatter(x_train, y_train, c='orange')
plt.plot(x_train, y_predict, c='blue')

plt.show()
```
