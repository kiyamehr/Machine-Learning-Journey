# Practice #1 Horse Power Price

```python
# Horse Power Price Prediction

# Horse Power(hp)
x_train = np.array([100.0, 150.0, 200.0, 250.0, 300.0])
# Price (1000$)
y_train = np.array([8.0, 12.0, 15.0, 18.0, 20.0])
```

```python
plt.scatter(x_train, y_train, marker='x')

plt.show()
```

```python
def compute_price(x ,w, b):
    m = x.shape[0]
    f_wb = np.zeros(m)
    
    for i in range(m):
        f_wb[i] = (w * x[i] + b)
        
    return f_wb
```

```python
tmp_f_wb = compute_price(x_train, 0.064, 1.8)
plt.plot(x_train, tmp_f_wb, c='green')
plt.scatter(x_train, y_train, marker="x")

plt.show()
```
