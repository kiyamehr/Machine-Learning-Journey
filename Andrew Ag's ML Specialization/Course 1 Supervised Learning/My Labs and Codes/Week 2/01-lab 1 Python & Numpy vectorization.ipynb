```python
# np.dot()

def my_dot (a, b):
    x = 0
    for i in range(a.shape[0]):
        x += a[i] * b[i]
    return x
```

```python
a = np.array([1, 2, 3, 4])
b = np.array([-1, 4, 3, 2])

print(my_dot(a, b))

print(np.dot(a, b))

#? why not do this too? (cause its less efficent)
# print(np.sum(a * b))
```
