```python
import numpy as np
import sklearn as skl
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('./Data/houses2features.csv').to_numpy()

print(data)
```

```python
# x & y must have the same Rows
x = np.take(data, indices=[0, 1], axis=1).reshape(-1, 2)
y = np.take(data, indices=2, axis=1).reshape(-1, 1)
```

```python
scaler = StandardScaler().fit(x)

x_scaled = scaler.transform(x)
print(x_scaled)
```

```python
plt.scatter(x[:, 0], y)
```

```python
regr = LinearRegression()

# Tries to find w & b for wx + b
regr.fit(x_scaled, y)
print(regr.score(x, y))
```

```python
y_pred = regr.predict(x_scaled)
y_pred
```

```python
mse = mean_squared_error(y, y_pred)
print(mse)
```

```python
plt.scatter(x[:, 0], y, c='blue', marker='x')
plt.scatter(x[:, 0], y_pred, c='red', marker='x')
```
