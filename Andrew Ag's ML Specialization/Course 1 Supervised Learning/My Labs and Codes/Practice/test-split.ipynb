```python
import numpy as np
import sklearn as skl
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('./Data/houses.csv').to_numpy()
print(data)
```

```python
x = np.take(data, indices=0, axis=1)
y = np.take(data, indices=1, axis=1)
print(x,'\n', y)
```

```python

split_point = int(x.shape[0] * 0.8)
```

```python
x_train = x[:split_point].reshape(-1, 1)
x_test = x[split_point:].reshape(-1, 1)

print(x_train)
print(x_test)
```

```python
y_train = y[:split_point].reshape(-1, 1)
y_test = y[split_point:].reshape(-1, 1)

print(y_train)
print(y_test)
```

```python
plt.scatter(y, x, c='green', marker='x')
plt.show()
```

```python
regr = LinearRegression()

regr.fit(x_train, y_train)

print(regr.score(x_train, y_train))
```

```python
y_test_predic = regr.predict(x_test)
y_train_predic = regr.predict(x_train)
```

```python
mse = mean_squared_error(y_test, y_test_predic)

print(mse)
```

```python
plt.scatter(x_test, y_test, c="orange")
plt.scatter(x_train, y_train, c="green")

plt.plot(x_test, y_test_predic, c="cyan")
plt.plot(x_train, y_train_predic, c="brown")
```
