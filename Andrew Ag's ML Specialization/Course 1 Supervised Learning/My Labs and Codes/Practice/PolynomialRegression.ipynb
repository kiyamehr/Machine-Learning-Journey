```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
data = pd.read_csv('./Data/polynomial.csv').to_numpy()
```

```python
x_train = np.take(data, indices=0, axis=1).reshape(-1, 1)
y_train = np.take(data, indices=1, axis=1).reshape(-1, 1)
```

```python
plt.scatter(x_train, y_train, marker="x")
```

```python
regr = LinearRegression()

regr.fit(x_train, y_train)
print(regr.score(x_train, y_train))
```

```python
y_predic = regr.predict(x_train)
```

```python
mse = mean_squared_error(y_train, y_predic)
print(mse)
```

# Doesnt fit really good

```python

plt.scatter(x_train, y_train, marker="x")
plt.plot(x_train, y_predic, c='green')
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_features = PolynomialFeatures(degree=2, include_bias=False)
x_poly = poly_features.fit_transform(x_train)
print(x_poly)
```

```python
curve_model = LinearRegression()
curve_model.fit(x_poly, y_train)
print(curve_model.score(x_poly, y_train))
```

```python
y_predic = curve_model.predict(x_poly)
```

```python
mse = mean_squared_error(y_predic, y_train)
print(mse)
```

```python
plt.scatter(x_train, y_predic, c='red')
plt.scatter(x_train, y_train, marker="x")
```
