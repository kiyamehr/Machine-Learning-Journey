```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

import numpy as np
```

```python
x_train = np.array([[2, 3]])
```

```python
model = Sequential([
    Dense(1, input_dim = 2, activation='sigmoid', name='L1')
])
```

```python
model.summary()
```

```python
layer_1 = model.get_layer('L1')
w, b =layer_1.get_weights()

print(f'w: {w}, b: {b}')
```

```python
y_pred1 = model.predict(x_train)

y_pred1
```

```python
set_w = np.array([[0.5], [-1]])
set_b = np.array([2.0])

model.layers[0].set_weights([set_w, set_b])

w, b = model.get_weights()

print(f'w: {w}, b: {b}')
```

```python
y_pred2 = model.predict(x_train)

y_pred2
```
