```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
X = np.array(
    [
        [1, 4],
        [2, 5],
        [2, 3],
        [3, 4],
        [4, 5],
        [4, 3],
        [5, 6],
        [5, 4],
        [6, 5],
        [7, 6],
        [7, 4],
        [8, 7],
    ],
    dtype=np.float32,
)

Y = np.array([0, 0, 0, 0, 1, 0, 1, 1, 1, 1, 1, 1], dtype=np.float32)
```

```python
norm_l = tf.keras.layers.Normalization(axis=-1)
norm_l.adapt(X)
Xn = norm_l(X)

print(Xn, Y)
print(Xn.shape, Y.shape)
```

```python
Xt = np.tile(Xn, (1000, 1))
Yt = np.tile(Y, (1000))

print(Xt.shape, Yt.shape)
```

```python
print(Xt.shape)
print(Yt.shape)
```

```python
tf.random.set_seed(1234)
model = Sequential(
    [
        tf.keras.Input(shape=(2,)), # Number of Features
        Dense(3, activation="sigmoid", name="L1"),
        Dense(1, activation="sigmoid", name="L2"),
    ]
)
```

```python
model.summary()
```

```python
w1, b1 = model.layers[0].get_weights() # method 1 to get layer
w2, b2 = model.get_layer('L2').get_weights() # method to get layer

print(w1.shape, b1.shape)
print(w1, "\n", b1)

print(w2.shape, b2.shape)
```

```python
model.compile(
    loss=tf.keras.losses.BinaryCrossentropy(),
    optimizer = tf.keras.optimizers.Adam(learning_rate=0.01)
    )

model.fit(
    Xt, Yt,
    epochs=10,
)
```

```python
print(f"layer 1, Old Weight: {w1}, \n old Bias: {b1}")
print(f"\n layer2, Old Weight: {w2}, old Bias: {b2}")

w_trained1, b_trained1 = model.layers[0].get_weights()
w_trained2, b_trained2 = model.layers[1].get_weights()

print(f"\n \n layer 1, Trained Weight: {w_trained1}, \n trained Bias: {b_trained1}")
print(f"\n layer2, Trained Weight: {w_trained2}, trained Bias: {b_trained2}")
```

```python

x_test = np.array([[2, 1], [3, 8]])

x_test_norm = norm_l(x_test)

print(x_test_norm)
```

```python
y_preds = model.predict(x_test_norm)

yhat = (y_preds >= 0.5).astype(int)

print(yhat)
```
