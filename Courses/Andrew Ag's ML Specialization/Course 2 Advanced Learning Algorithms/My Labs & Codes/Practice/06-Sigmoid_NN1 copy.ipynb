```python
import numpy as np
import tensorflow as tf
from keras import layers
```

```python
x_train = np.array([[2, 8], [3, 7], [4, 7], [5, 6], [6, 6], [7, 5], [8, 5], [1, 9]])
y_train = np.array([0, 0, 0, 1, 1, 1, 1, 0])

print(x_train.shape)
print(y_train.shape)
```

```python
x_train = x_train / np.max(x_train, axis=0)
print(x_train)
```

```python
tf.random.set_seed(1234)
model = tf.keras.Sequential(
    [
        layers.Dense(2, activation="relu", name="L1"),
        layers.Dense(1, activation="sigmoid", name="L2"),
    ]
)
```

```python
model.compile(
    loss=tf.keras.losses.BinaryCrossentropy(),
    optimizer=tf.keras.optimizers.Adam(1.0))

model.fit(x_train, y_train, epochs=100)
```

```python
x_test = np.array([[2, 5]])
y_predic = model.predict(x_test)
y_predic
```
