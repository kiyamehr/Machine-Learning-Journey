```python
import numpy as np
```

```python
x = np.array([2, 3])
```

```python
w = np.array([0.5, -1])
b = 2
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
     
def neuron(x, w, b):

    z = np.dot(w, x) + b

    pred = sigmoid(z)

    if pred >= 0.5:
        return f"Number: {pred} was predicted as 1"
    else:
        return f"Number: {pred} was predicted as 0"
```

```python
prediction = neuron(x, w, b)

print(prediction)
```
