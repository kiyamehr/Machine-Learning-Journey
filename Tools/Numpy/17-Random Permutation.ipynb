```python
from numpy import random 
import numpy as np
```

```python
arr = np.array([1, 2, 3, 4, 5])

# Changes the Original Array
random.shuffle(arr)

print(arr)
```

```python
arr = np.array([1, 2, 3, 4, 5])

# Leaves the Original Array unchanged
print(random.permutation(arr))
```
