# Data types in Numpy

### Below is a list of all data types in NumPy and the characters used to represent them.

- i - integer
- b - boolean
- u - unsigned integer
- f - float
- c - complex float
- m - timedelta
- M - datetime
- O - object
- S - string
- U - unicode string
- V - fixed chunk of memory for other type ( void )


```python
import numpy as np
```

```python
arr = np.array([1, 2, 3, 4])

print(arr.dtype)
```

```python
arr = np.array(["apple", "banana", "cherry"])

print(arr.dtype)
```

```python
arr = np.array([1, 2, 3, 4], dtype='S')

print(arr)
print(arr.dtype)
```

```python
# gets stored in 8-bit data so 4*8 = 32
arr = np.array([1, 2, 3, 4], dtype="i4")

print(arr)
print(arr.dtype)
```

```python
# raises Error
arr = np.array(["a", "2", "3"], dtype="i")
```

```python
arr = np.array([1.1, 2.1, 3.1])

newArr = arr.astype('i')

print(newArr)
print(newArr.dtype)
```

```python
arr = np.array([1.1, 2.1, 3.1])

newArr = arr.astype(int)

print(newArr)
print(newArr.dtype)
```

```python
arr = np.array([1, 2, 0, 3, 0])

newArr = arr.astype(bool)

print(newArr)
print(newArr.dtype)
```
