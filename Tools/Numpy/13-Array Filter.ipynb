# But the Better Example is:

````python
arr = np.array([41, 42, 43, 44])

filter_arr = []

for element in arr:
    if element > 42:
        filter_arr.append(True)
    else:
        filter_arr.append(False)

print(arr[filter_arr])
````

````python
arr = np.array([41, 42, 43, 44])

filter_arr = []

for index, element in np.ndenumerate(arr):
    if arr[index] == 41 or arr[index] == 43:
        filter_arr.append(True)

    else:
        filter_arr.append(False)

print(filter_arr)
print(arr[filter_arr])
````

````python
arr = np.array([41, 42, 43, 44])

filter_arr = arr > 42

print(arr[filter_arr])
````
