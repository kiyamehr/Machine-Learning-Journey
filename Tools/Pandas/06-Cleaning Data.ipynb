# Calculating empty cells with Mean
### The average value (the sum of all values divided by number of values).

```python
df = pd.read_csv("./Pandas Data/workout_data.csv")

mean_calories = df["Calories"].mean()

df.fillna({"Calories": mean_calories}, inplace=True)

print(df.to_string())
```

# with Median
### the value in the middle, after you have sorted all values ascending.

```python
df = pd.read_csv("./Pandas Data/workout_data.csv")

df_median = df["Calories"].median()
df_mean = df["Calories"].mean()

df.fillna({"Calories": df_median}, inplace=True)

print(df_median)
print(df)
```

# with Mode
### the value that appears most frequently.

```python
df = pd.read_csv("./Pandas Data/workout_data.csv")

df_mode = df["Calories"].mode()

df.fillna({"Calories": df_mode}, inplace=True)

print(df_mode)
print(df)
```

# Fixing Wrong Format -------------------

```python
df = pd.read_csv("./Pandas Data/workout_data.csv")

df["Date"] = pd.to_datetime(df["Date"], format="mixed")
df.dropna(subset="Date", inplace=True)

print(df.to_string())
```

# Wrong Data -----------------------

```python
df = pd.read_csv("./Pandas Data/workout_data.csv")

# df.loc[7, 'Duration'] = 45


for x in df.index:
    if df.loc[x, "Duration"] > 120:
        df.loc[x, "Duration"] = 120
        
        # removing the wrong data
        df.drop(x, inplace=True)
    

print(df.to_string())
```

# Duplicated

```python
def calc_mean(column):
    if df[column].dtype == "int64":
        return df[column].mean()
    else:
        print("not int")

df = pd.read_csv("Pandas Data/workout_data.csv")

duration_mean = df["Duration"].mean()
pulse_mean = df["Pulse"].mean()
maxpulse_mean = df["Maxpulse"].mean()
calories_mean = df["Calories"].mean()

df.fillna({"Duration": duration_mean}, inplace=True)
df.fillna({"Pulse": pulse_mean}, inplace=True)
df.fillna({"Maxpulse": maxpulse_mean}, inplace=True)
df.fillna({"Calories": calories_mean}, inplace=True)

df['Date'] = pd.to_datetime(df['Date'], format='mixed')
df.dropna(subset='Date', inplace=True)

for x in df.index:
    if df.loc[x, 'Duration'] > 120:
        df.drop(x, inplace=True)

print(df.to_string())
```
